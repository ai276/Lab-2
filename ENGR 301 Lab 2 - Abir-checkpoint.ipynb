```python
# H0: Blood Pressure is from a normal distribution
# H1: Blood Pressure is not from a normal distribution

stats.normaltest(df["BloodPressure"])

# p<alpha
# H0 is rejected, while H1 is accepted
```

The null hypothesis states that the data comes from a normal distribution whereas the alternate hypothesis says it does not. The pvalue is 2.7337301565003257e-05 which means it is not normally distributed siding with the alternative hypothesis.

Problem 3:

```python
# Part a

# H0: mean of Blood Pressure of those without Diabetes is almost similar to the mean of the Blood Pressure for the whole population
# H1: mean of Blood Pressure of those without Diabetes is not similar to the mean of the Blood Pressure for the whole population

stats.ttest_1samp( df[df["Outcome"] == 0]["BloodPressure"], 71)

# p>0.01
# Accepts H0
```

No it is not significantly different.

```python
# Part b

# H0: mean of Blood Pressure of those without Diabetes is almost similar to the mean of the Blood Pressure for the whole population
# H1: mean of Blood Pressure of those without Diabetes is not similar to the mean of the Blood Pressure for the whole population

stats.ttest_1samp( df[df["Outcome"] == 1]["BloodPressure"], 71)

# p>0.01
# Accepts H0
```

Yes it is significantly different.

```python
# Part c

# H0: mean of Blood Pressure of those with Diabetes is similar to the mean of those without it
# H1: mean of Blood Pressure of those with Diabetes is no similar to the mean of those without it

stats.ttest_ind( df[df["Outcome"] == 1]["BloodPressure"], df[df["Outcome"] == 0]["BloodPressure"])

# p<0.01
# H0 is rejected
```

Yes it is significantly different.
